```python
import numpy as np
import scipy.stats as si

S = 90
K = 100
r = 0.01
sigma = 0.3
T_list = [1, 0.9, 0.5, 0.3, 0.1]

print("S, K, r, sigma =", S, K, r, sigma)
for T in T_list:
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    
    call_price = S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0)
    
    print(f"T = {T:.1f}, Call Price = {round(call_price, 4)}")
import matplotlib.pyplot as plt
T_list = [1, 0.9, 0.5, 0.3, 0.1]
call_prices = [7.3253,6.7502,4.1281,2.5242,0.6167]
plt.figure(figsize=(8, 5))
plt.plot(T_list ,call_prices , label='Call Option Price', color='blue')
plt.title('European Call Option Price vs. Time to Maturity')
plt.xlabel('Time to Maturity (T)')
plt.ylabel('Call Option Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
